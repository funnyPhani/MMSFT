```python
import matplotlib.pyplot as plt
import numpy as np

# Provided data
monolingual_scores = [28.22, 29.39, 29.24]
multilingual_scores = [30.51, 31.07, 32.55]
scenarios = ['mT5', 'VG-mT5', 'MMSFT']
hatch_patterns = ['//', '++', 'xx']

# Check if the lengths match
if len(scenarios) != len(hatch_patterns):
    raise ValueError("Mismatch between the number of scenarios and hatch patterns")

# Set up positions for bars
x = np.arange(len(scenarios))

# Plotting
bar_width = 0.25  # Reduced bar width
plt.bar(x - bar_width / 2, monolingual_scores, width=bar_width, label='Monolingual', hatch=hatch_patterns[0])
plt.bar(x + bar_width / 2, multilingual_scores, width=bar_width, label='Multilingual', hatch=hatch_patterns[1])

# Adding labels on top of the bars
for i, (monolingual, multilingual) in enumerate(zip(monolingual_scores, multilingual_scores)):
    plt.text(i - bar_width / 2, monolingual + 0.5, f'{monolingual:.2f}', ha='center', va='bottom')
    plt.text(i + bar_width / 2, multilingual + 0.5, f'{multilingual:.2f}', ha='center', va='bottom')

# Adding labels and title
plt.xlabel('Scenarios')
plt.ylabel('Average LaTM Scores')
plt.title('Avg. LaTM Scores for Monolingual and Multilingual Scenarios of low-resources languages')
plt.xticks(x, scenarios)
plt.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Provided data
monolingual_scores = [28.22, 29.39, 29.24]
multilingual_scores = [30.51, 31.07, 32.55]
scenarios = ['mT5', 'VG-mT5', 'MMSFT']
hatch_patterns = ['//', '++', 'xx']

# Check if the lengths match
if len(scenarios) != len(hatch_patterns):
    raise ValueError("Mismatch between the number of scenarios and hatch patterns")

# Set up positions for bars
x = np.arange(len(scenarios))

# Plotting
bar_width = 0.35  # Reduced bar width
plt.bar(x - bar_width / 2, monolingual_scores, width=bar_width, label='Monolingual', hatch=hatch_patterns[0])
plt.bar(x + bar_width / 2, multilingual_scores, width=bar_width, label='Multilingual', hatch=hatch_patterns[1])

# Adding labels on top of the bars
for i, (monolingual, multilingual) in enumerate(zip(monolingual_scores, multilingual_scores)):
    plt.text(i - bar_width / 2, monolingual + 0.5, f'{monolingual:.2f}', ha='center', va='bottom', weight='bold')
    plt.text(i + bar_width / 2, multilingual + 0.5, f'{multilingual:.2f}', ha='center', va='bottom', weight='bold')

# Adding labels and title with bold text
# plt.xlabel('Scenarios', weight='bold')
plt.ylabel('Average LaTM Scores', weight='bold')
# plt.title('Avg. LaTM Scores for Monolingual and Multilingual Scenarios of low-resources languages', weight='bold')
plt.xticks(x, scenarios)
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1), fancybox=True, shadow=True, ncol=2)
# Show the plot
plt.savefig('comparison_plot_L.png', bbox_inches='tight',dpi=500)
plt.show()
```
